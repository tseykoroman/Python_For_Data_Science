### Задание 2
С помощью функции linspace из библиотеки Numpy создайте массив t из 51 числа от 0 до 10 включительно.
Создайте массив Numpy под названием f, содержащий косинусы элементов массива t.
Постройте линейную диаграмму, используя массив t для координат по горизонтали,а массив f - для координат по вертикали. Линия графика должна быть зеленого цвета.
Выведите название диаграммы - 'График f(t)'. Также добавьте названия для горизонтальной оси - 'Значения t' и для вертикальной - 'Значения f'.
Ограничьте график по оси x значениями 0.5 и 9.5, а по оси y - значениями -2.5 и 2.5.

```python
t = np.linspace(0, 10, 51)
f = np.cos(t)


legend_font = {
    "size": 7,
    "family": "serif",
}

label_font = {
    "fontsize": 9,
    "family": "serif",
}

title_font = {
    "fontsize": 16,
    "fontweight": "bold",
    "color": "#808080",
    "family": "serif",
}

plt.plot(t, f, label="cos(t)", color="green", linestyle="dashed")

plt.title("График f(t)", fontdict=title_font)
plt.xlabel("Значения t", fontdict=label_font)
plt.ylabel("Значения f", fontdict=label_font)

plt.axis([0.5, 9.5, -2.5, 2.5])

legend = plt.legend(loc="lower right", prop=legend_font)
plt.setp(legend.get_texts(), color="DarkBlue")

plt.grid(axis="x", color="lightgrey")

plt.show()
```

С помощью функции linspace библиотеки Numpy создайте массив x из 51 числа от -3 до 3 включительно.
Создайте массивы y1, y2, y3, y4 по следующим формулам:
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = sin(x)
Используя функцию subplots модуля matplotlib.pyplot, создайте объект matplotlib.figure.Figure с названием fig и массив объектов Axes под названием ax,причем так, чтобы у вас было 4 отдельных графика в сетке, состоящей из двух строк и двух столбцов. 

В каждом графике массив x используется для координат по горизонтали. В левом верхнем графике для координат по вертикали используйте y1,в правом верхнем - y2, в левом нижнем - y3, в правом нижнем - y4. Дайте название графикам: 'График y1', 'График y2' и т.д.
Для графика в левом верхнем углу установите границы по оси x от -5 до 5.
Установите размеры фигуры 8 дюймов по горизонтали и 6 дюймов по вертикали.
Вертикальные и горизонтальные зазоры между графиками должны составлять 0.3.


```python
x = np.linspace(-3, 3, 51)
y1 = x ** 2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)

title_font = {
    "fontsize": 16,
    "fontweight": "bold",
    "color": "#808080",
    "family": "serif",
}

fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax1.plot(x, y1)
ax1.set_xlim([-5, 5])
ax1.set_title("График y1", fontdict=title_font)

ax2.plot(x, y2)
ax2.set_title("График y2", fontdict=title_font)

ax3.plot(x, y3)
ax3.set_title("График y3", fontdict=title_font)

ax4.plot(x, y4)
ax4.set_title("График y4", fontdict=title_font)
```

Получить матрицу B, умножив матрицу A на матрицу At с помощью матричного умножения. Какой размер имеет матрица B? Получится ли вычислить обратную матрицу для матрицы B и почему?


```python
a = np.arange(0, 12)
print('Одномерный массив а', a, sep='\n')
A = a.reshape(4, 3)
print('Двумерный массив Numpy - матрица A', A, sep='\n')
At = A.transpose()
print('Матрица At', At, sep='\n')
B = np.dot(A, At)
print('Матрица B', B, sep='\n')
print('При умножении матриц число строк итоговой матрицы (B) равно число строк первой матрицы (A), а число столбцов - второй матрицы (At).', sep='\n')
print('Размер матрицы B', B.size, sep='\n')
print('Вычислить обратную матрицу для B невозможно, т.к. определитель матрицы B равен 0.', sep='\n')
print('Определитель матрицы B', np.linalg.det(B), sep='\n')
```
